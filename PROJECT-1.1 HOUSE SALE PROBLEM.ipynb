```python
import pandas as pd
import numpy as np
import xlrd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Importing Data and Testing Files

```python
htrain = pd.read_csv('F:/DATA SCIENCE/PROJECT 1 HOUSE PRICES/train.csv')
htest = pd.read_csv('F:/DATA SCIENCE/PROJECT 1 HOUSE PRICES/test.csv')

#The above code loads Data File(train) and Test File (test) into the Jupyter.
```

# Preparing Data File 

```python
h_train_num=htrain.select_dtypes(include=[np.number])
#Filters Numericals from the overall data
h_train_cat=htrain.select_dtypes(include=['object'])
#Filters Categorical/Words from the overall data
```

```python
h_train_num.columns
```

```python
h_train_cat.columns
```

```python
for HC in h_train_num.columns:
    print (HC)
    print (h_train_num[HC].mean())
    print (h_train_num[HC].median())
    print('--------------------------------')
    
# Created a Data Frame 'HC'. Calculated mean and median of h_train_num and filled in HC
# in Ideal conditions, mean and medians are same in each case.
```

```python
h_train_num_corr=h_train_num.corr()
# Finds out co-relation between the columns in the data in order to find the important features
```

```python
h_train_num_corr["SalePrice"]
# Gives the relation of Year Built wrt other columns
```

```python
h_train_num_corr[(h_train_num_corr["SalePrice"]>0.35) ]
# Gives the complete data of relationships which are greater than 0.30
```

```python
h_train_num_corr[(h_train_num_corr["SalePrice"]<-0.35) ]
# Gives the complete data of relationships which are less than 0.3 (its our wish), Usually it is between 0.3 to 0.6
```

```python
h_train_num_corr[(h_train_num_corr["SalePrice"]>0.35) ].index.values
# This gives the details of those columns which have the co-relation value between -1 to 0.35 and 0.35 to 1 
```

```python
h_train_num_corr[(h_train_num_corr["SalePrice"]<-0.35) ].index.values
# To choose columns whose relationship lies between -1 to -0.35 (unfortunately there are no outputs)
```

```python
h_train_num_cols = []
h_train_num_cols.extend(h_train_num_corr[(h_train_num_corr["SalePrice"]>0.35) ].index.values)
h_train_num_cols.extend(h_train_num_corr[(h_train_num_corr["SalePrice"]<-0.35) ].index.values)

# Created a Data Frame h_train_num_cols and imported values of sales price relations
```

```python
h_train_num_cols
```

```python
h_train_num_col_filtered=h_train_num[h_train_num_cols]

# From overall data( i.e. h_train_num) we are choosing only those columns which are effecting more(i.e. h_train_num_cols) as per the correlation resuts
```

```python
h_train_num_col_filtered
```

```python
(h_train_num_col_filtered.isnull().sum().sort_values(ascending=False))
# Finding the empty cells
```

```python
for col in ["LotFrontage","GarageYrBlt","MasVnrArea"]:
    h_train_num_col_filtered[col].fillna(h_train_num_col_filtered[col].median(),inplace=True)
    # This step fills Nans with median value
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
# Fro here we start working on categoricals 
```

```python
h_train_cat.head()
```

```python
(h_train_cat.isnull().sum().sort_values(ascending=False))
#To find out NaNs in the categoricals
```

```python
for col in ["PoolQC","MiscFeature","Alley","Fence","FireplaceQu"]:
    h_train_cat[col].fillna('No Value',inplace=True)
# Here we are filling NaNs in "PoolQC","MiscFeature","Alley","Fence","FireplaceQu" with No Value Since more than half data is missing
```

```python
for col in ["GarageCond","GarageQual","GarageFinish","GarageType","BsmtFinType2","BsmtExposure","BsmtFinType1","BsmtQual","BsmtCond","MasVnrType","Electrical"]:
    h_train_cat[col].fillna(h_train_cat[col].value_counts().idxmax(),inplace=True)
    # This step is to fill the NaNs with Mode values
```

```python
h_train_cat.head()
```

```python
h_train_cat1=h_train_cat.apply(le.fit_transform)
# This steps encodes the words with a code(numerical)
```

```python
h_train_cat1.head()
```

```python
TrainingDATA=pd.concat([h_train_num_col_filtered,h_train_cat1],axis=1)
```

```python
TrainingDATA.head()
```

## Preparing Test Data File

```python
h_test_num=htest.select_dtypes(include=[np.number])
#Filters Numericals from the overall data

h_test_cat=htest.select_dtypes(include=['object'])
#Filters Categorical/Words from the overall data
```

```python
h_train_num_cols.remove('SalePrice')
# Removes SalePrice from Training Data and Keeps the remaining Data COlumns
# Note: The Data thats is in Training should be in Testing as well
```

```python
h_test_num_col_filtered=h_test_num[h_train_num_cols]
```

```python
(h_test_num_col_filtered.isnull().sum().sort_values(ascending=False))
# Finding the Nans
```

```python
for hc in ["LotFrontage","GarageYrBlt","MasVnrArea","TotalBsmtSF","GarageArea","GarageCars","BsmtFinSF1"]:
    print (hc)
    print(h_test_num_col_filtered[hc].mean())
    print(h_test_num_col_filtered[hc].median())
    print ("------------------------------")
```

```python
for col in ["LotFrontage","GarageYrBlt","MasVnrArea","TotalBsmtSF","GarageArea","GarageCars","BsmtFinSF1"]:
    h_test_num_col_filtered[col].fillna(h_test_num_col_filtered[col].median(),inplace=True)
```

```python
(h_test_cat.isnull().sum().sort_values(ascending=False))
# Fining the NaNs in Categoricals
```

```python
for col in ["PoolQC","MiscFeature","Alley","Fence","FireplaceQu"]:
    h_test_cat[col].fillna('No Value',inplace=True)
```

```python
for col in ["GarageCond","GarageQual","GarageFinish","GarageType","BsmtQual","BsmtCond","BsmtFinType2","BsmtExposure","BsmtFinType1","MasVnrType","MSZoning","Utilities","Functional","KitchenQual","SaleType","Exterior2nd","Exterior1st"]:
    h_test_cat[col].fillna(h_test_cat[col].value_counts().idxmax(),inplace=True)
```

```python
h_test_cat1=h_test_cat.apply(le.fit_transform)
```

```python
TestingDATA=pd.concat([h_test_num_col_filtered,h_test_cat1],axis=1)
```

```python
TestingDATA.head()
```

# Splitting Train File into Training Dataset and Testing Dataset to Build Models

```python
from sklearn.model_selection import train_test_split
```

```python
XX=TrainingDATA.drop(["SalePrice"],axis=1)
```

```python
YY=TrainingDATA["SalePrice"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(XX, YY, test_size=0.3)

#X_Train will consist of 70% data and X Test will Consist of 30% data for Testing
# X = all the columns except output(Salesprice)
# Y = only SalesPrice that has to be predicted
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# LINEAR REGRESSION MODEL SCORE


```python
from sklearn.linear_model import LinearRegression
```

```python
Linreg=LinearRegression()
```

```python
lm_M=Linreg.fit(X_train,y_train)
```

```python
lm_M.score(X_train,y_train)
```

# DECISION TREE REGRESSION MODEL SCORE

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DecTree=DecisionTreeRegressor()
```

```python
dt_M=DecTree.fit(X_train,y_train)
```

```python
dt_M.score(X_train,y_train)
```

# RANDOM FOREST REGRESSION MODEL SCORE

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RandFor=RandomForestRegressor(n_estimators=5000)
```

```python
rf_M=RandFor.fit(X_train,y_train)
```

```python
rf_M.score(X_train,y_train)
```

# GRADIENT BOOST REGRESSION MODEL

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
GBM=GradientBoostingRegressor(n_estimators=3000)
```

```python
gb_M=GBM.fit(X_train,y_train)
```

```python
gb_M.score(X_train,y_train)
```

# TEST SCORES OF MODELS BUILT

```python
from sklearn.metrics import accuracy_score, r2_score
```

```python
gb_y_pred = gb_M.predict(X_test)
rf_y_pred = rf_M.predict(X_test)
dt_y_pred = dt_M.predict(X_test)
lm_y_pred = lm_M.predict(X_test)
```

```python
r2_score(y_test,gb_y_pred)
```

```python
r2_score(y_test,rf_y_pred)
```

```python
r2_score(y_test,dt_y_pred)
```

```python
r2_score(y_test,lm_y_pred)
```

# Since the r2_score of Gradient Boost is better than other algorithms, the  actual Testing File is now executed using Graident model to predict the Sales Price.

```python
y_eval=gb_M.predict(TestingDATA)
# Predicts the SalesPrice
```
